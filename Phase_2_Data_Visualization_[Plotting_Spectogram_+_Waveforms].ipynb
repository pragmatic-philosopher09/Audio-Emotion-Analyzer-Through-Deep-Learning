# **Preprocessing of Audio files**



>This block of code is useful for visualization of audio files in .wav format.
The main packages used are ffmeg,wav_read
This code line effectively updates the chunk size field in the RIFF chunk with the correct value, which is needed because the chunk size field may not be updated automatically when the audio data changes. 

```python
import IPython.display as ipd
import numpy as np
import scipy.io.wavfile as wav
import io
import ffmpeg

import librosa
import librosa.display
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
```

```python
def get_audio(filepath):
    binary = io.open(filepath, 'rb').read()
    process = (ffmpeg
        .input('pipe:0')
        .output('pipe:1', format='wav')
        .run_async(pipe_stdin=True, pipe_stdout=True, pipe_stderr=True, quiet=True, overwrite_output=True)
    )
    output, err = process.communicate(input=binary)

    sr, audio = wav.read(io.BytesIO(output))
    
    return audio, sr

filepath = '5 Sec.mp3'
audio, sr = get_audio(filepath)
```

**The resulting spectrogram is displayed using the librosa.display.specshow() function with a logarithmic frequency axis (y_axis='log') and a linear time axis (x_axis='time'). The librosa.amplitude_to_db() function is used to convert the amplitude scale to decibels (dB), which is a more convenient scale for visualizing audio signals.**


```python
spec = np.abs(librosa.stft(a1, hop_length = 512))
spec = librosa.amplitude_to_db(spec, ref = np.max)

plt.figure(figsize=(20, 10))
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format = '%+2.0f dB');
plt.title("Spectogram");
     
```

# **Mel Spectogram**

```python
mel_spec = librosa.feature.melspectrogram(y=a1, sr=sr1, hop_length = 1024)
mel_spec = librosa.power_to_db(mel_spec, ref = np.max)

plt.figure(figsize=(20, 10))
librosa.display.specshow(mel_spec, fmax=8000, x_axis='time', y_axis='mel');
plt.colorbar(format = '%+2.0f dB');
plt.title("Mel Spectogram");
     
```

# **Spectral Centroid**

*1. Spectral centroid is a particular form of measure of audio feature.*

*2. It seeks to represent the "centre of mass/gravity" of audio, by computing the **weighted mean of frequencies** depicted in the frequency domain of the audio in question.*

*3. The spectrum of sound signal takes a **specific number of frequency bins** generated **by STFT**, which is used as a measure of computing the spectral centroid.*

```python
spectral_centroids1 = librosa.feature.spectral_centroid(a1,sr=sr1)[0]
spectral_centroids1.shape
```

### ***Spectral Centroid for "Neutral" Male Emotion***

```python
plt.figure(figsize=(20,5))
frames = range(len(spectral_centroids1))
t = librosa.frames_to_time(frames)

def normalize(x, axis=0):
  return sklearn.preprocessing.minmax_scale(x,axis=axis)

librosa.display.waveplot(a1, sr=sr1, alpha = 1)

plt.plot(t, normalize(spectral_centroids1), color='#0bbf26')
```

# **Spectral Rolloff**

*1. Spectral rolloff is a measure of the shape of the audio signal.*

*2. Percentage below which a specified frequency of the total spectral energy lies.*

```python
plt.figure(figsize=(20,5))
spectral_rolloff = librosa.feature.spectral_rolloff(a1+0.01, sr=sr1)[0]

librosa.display.waveplot(a1,sr=sr1,alpha=0.5)
plt.plot(t,normalize(spectral_rolloff),color='#13a179')
plt.grid()
```
